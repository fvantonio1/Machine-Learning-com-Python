# Algoritmo Apriori

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
base_mercado1 = pd.read_csv('mercado.csv', header=None)
base_mercado1
```

```python
transacoes = []
for i in range(len(base_mercado1)):
  transacoes.append([str(base_mercado1.values[i,j]) for j in range(4)])
```

```python
transacoes
```

```python
regras = apriori(transacoes, min_support=0.3, min_confidence=0.8, min_lift=2)
resultados = list(regras)
```

```python
resultados
```

```python
resultados[2]
```

```python
r = resultados[2][2]
r
```

```python
se = []
entao = []

suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]

  regras = resultado[2]
  for regra in regras:
    se.append(list(regra[0]))
    entao.append(list(regra[1]))
    confianca.append(regra[2])
    suporte.append(s)
    lift.append(regra[3])
```

```python
rules_df = pd.DataFrame({'se': se, 'entao': entao, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
rules_df.sort_values(by='lift', ascending=False)
```

```python
base_mercado2 = pd.read_csv('mercado2.csv', header=None)
base_mercado2
```

```python
transacoes = []

for i in range(base_mercado2.shape[0]):
  transacoes.append([str(base_mercado2.values[i,j]) for j in range(base_mercado2.shape[1])])
```

```python
# produtos que são vendidos quatro vezes ao dia
# 7 * 4 = 28

28 / 7501
```

```python
regras = apriori(transacoes, min_support=0.0035, min_confidence=0.2, min_lift=3)
resultados = list(regras)
len(resultados)
```

```python
se = []
entao = []

suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]

  regras = resultado[2]
  for regra in regras:
    se.append(list(regra[0]))
    entao.append(list(regra[1]))
    confianca.append(regra[2])
    suporte.append(s)
    lift.append(regra[3])
```

```python
rules_df = pd.DataFrame({'se': se, 'entao': entao, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
rules_df.sort_values(by='confianca', ascending=False)
```
